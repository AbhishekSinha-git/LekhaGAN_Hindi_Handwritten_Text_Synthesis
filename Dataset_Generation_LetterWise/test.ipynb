```python
import numpy as np
import matplotlib.pyplot as plt
import random

# --- Config ---
npy_file_path = './LetterClassPairsCompressed/0939_094C_1000_images.npy'  # Update this path as needed
image_index = random.randint(0, 1000)  # Change this index to view different images
EXPECTED_HEIGHT = 64
EXPECTED_WIDTH = 256
# ---------------

# Load the .npy file
data = np.load(npy_file_path)

# Check shape info
print(f"Loaded shape: {data.shape}")  # Expect (rows, 64, 25600) if 100 images per row

# Extract the image from a row
row = data[image_index // 100]
img_start = (image_index % 100) * EXPECTED_WIDTH
img_end = img_start + EXPECTED_WIDTH
image = row[:, img_start:img_end]

# Sanity check
assert image.shape == (EXPECTED_HEIGHT, EXPECTED_WIDTH), f"Got unexpected shape: {image.shape}"

# Split into left and right halves
left = image[:, :128]
right = image[:, 128:]

# --- Display both halves ---
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(left, cmap='gray')
axes[0].set_title("Left Half (64x128)")
axes[0].axis('off')

axes[1].imshow(right, cmap='gray')
axes[1].set_title("Right Half (64x128)")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
